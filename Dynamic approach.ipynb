```python
def read_input(filename):
    lines = []

    with open(filename,'r') as conllfile:
        for line in conllfile:
            columns = line.split()
            lines.append(columns)

    return lines
```

```python
def clean_lines(raw_lines):
    lines_clean = []
    for line in raw_lines:
        if line == []:
            continue
        elif not line[0].startswith('/friends'):
            continue
        else:
            lines_clean.append(line)

    return lines_clean
```

```python
def prepare_timestamps(cleanlines):
    running_order = 0 

    for index in range(0,len(cleanlines)):
        season = cleanlines[index][0][11]
        episode = cleanlines[index][0][-2:]
        cleanlines[index][0] = season+"_"+episode
        if cleanlines[index-1][1] != cleanlines[index][1]:
            running_order = 0
        else:
            running_order += 1
        cleanlines[index][2] = running_order

    return cleanlines
```

```python
def prepare_data(filename):
    raw_data = read_input(filename)
    clean_data = clean_lines(raw_data)
    clean_data = prepare_timestamps(clean_data)
    
    return clean_data
```

```python
file = "friends.test.episode_delim.conll.txt"

working_data = prepare_data(file)
```

```python
bridge_start = []
bridge_end = []
single = []

for index, line in enumerate(working_data):
    if line[-1] == '-':
        continue
    elif not line[-1].endswith(')'):
        bridge_start.append(index)
    elif not line[-1].startswith('('):
        bridge_end.append(index)
    else:
        single.append(index)
    
ments_ordered = sorted(bridge_start+single)

    
    
```

## Get entities with mention

```python
references = []

for index in ments_ordered:
    if index in bridge_start:
        position = bridge_start.index(index)
        end = bridge_end[position]
        current_ref = []
        for pos in range(index, end+1):
            ref_part = working_data[pos][3]
            current_ref.append(ref_part)
        ref = ' '.join(current_ref)
    else:
        ref = working_data[index][3]
    references.append(ref)
    
print(references)
```

```python
ent_with_ref = list(zip(ents_from_text, references))

print(ent_with_ref)
```

```python
entity_dict = {}

with open('friends_entity_map.txt') as entity_map:
    for line in entity_map:
        nameslist = line.split()
        name = '_'.join(nameslist[1:])
        entity = nameslist[0]
        entity_dict[entity] = name
        
print(entity_dict)
        
```

```python
all_mentions= []

for ment_id in ments_ordered:
    se_ep = working_data[ment_id][0]
    scene = working_data[ment_id][1]
    ts = working_data[ment_id][2]
    timestamp = f'{se_ep}_{scene}_{ts}'
    position = ments_ordered.index(ment_id)
    entity = ent_with_ref[position][0]
    reference = ent_with_ref[position][1]
    name = entity_dict[entity]
    mention_dict = {'timestamp':timestamp, 'entity':entity, 'reference':reference, 'name':name}
    all_mentions.append(mention_dict)

print(all_mentions)
```

```python
all_timestamps = []

for ment_dict in all_mentions:
    all_timestamps.append(ment_dict['timestamp'])

print(all_timestamps)
```

## Categorisation of mentions into types

```python
unique_refs = list(set(references))
```

```python
import spacy

nlp = spacy.load('en_core_web_sm')

names = []
prons = []
other = []

for ref in unique_refs:
    doc = nlp(ref)
    postags = []
    for token in doc:
        postags.append(token.pos_)
    if postags[-1] == "PRON":
        prons.append(ref)
    elif postags[-1] == "PROPN":
        names.append(ref)
    else:
        other.append(ref)


names.remove('mine')
prons.append('mine')

other_to_names = ['Mindy', 'Todd', 'Geller', 'Silvian', 'Bookbinders', 'Paulo', 'Drake', 'Bing', 'Peanut', 'Ursula',
                 "Andie McDowell 's",'Gobb', "Ugly Naked Guy 's"]

for name in other_to_names:
    names.append(name)
    try:
        other.remove(name)
    except ValueError:
        print(name + "not in list")
    
fst_snd_prons = []

for pron in prons:
    if pron == 'I':
        fst_snd_prons.append(pron)
    elif pron.lower().startswith('m'):
        fst_snd_prons.append(pron)
    elif pron.lower().startswith('y'):
        fst_snd_prons.append(pron)
        
for pron in fst_snd_prons:
    prons.remove(pron)

other_to_fst_snd = ['Ya', "what'cha", 'Whaddya', 'howdya']

for fst_snd in other_to_fst_snd:
    fst_snd_prons.append(fst_snd)
    other.remove(fst_snd)

endearing = []
names_to_endearing = ['Nana', 'Mon', 'Rossy', 'Dad', 'pal', 'Mommy']

for dear in names_to_endearing:
    endearing.append(dear)
    names.remove(dear)
    
other_to_endearing = ['Phoebs', 'mom', 'buddy', 'Hon', 'Pheebs', 'dad', 'Urse', 'Daddy', 'birthday sweetie', 'honey', 'Rach']

for dear in other_to_endearing:
    endearing.append(dear)
    other.remove(dear)
    
names_to_other = ['buyer', 'Mr. Sweet']

for oth in names_to_other:
    other.append(oth)
    names.remove(oth)
    
errors = ['her I', 'him I']

for err in errors:
    prons.remove(err)

context_dep = other

print(names)
print(fst_snd_prons)
print(prons)
print(endearing)
print(context_dep)
print(errors)
```
